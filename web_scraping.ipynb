```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
```

# set the url

```python
url = "https://www.bikewale.com/royalenfield-bikes"
```

```python
page=requests.get(url)
```

```python
page
```

```python
#Now Scraping the things using the Beautiful soup in Python
```

# soup object


```python
soup=BeautifulSoup(page.content,"html.parser")
```

```python
print(soup.text)
```

```python
image=soup.find_all("div",class_="imageWrapper")
```

```python
len(image)
```

```python
print(image)
```

```python
for i in image:
    j=i.img["src"]
    print(j)
    




```

```python
#storing all the images
img1=[]
for i in image:
    j=i.img["src"]
    img1.append(j)
print(img1)
    


```

```python
#updated images 
royal_image=img1[:7]
len(royal_image)
```

```python
#getting the links
links=[]
link=soup.findAll("div",class_="bikeDescWrapper")
for i in link:
    j=i.a["href"]
    links.append(j)
print(links)
```

```python
link_1=[]
link=soup.findAll("div",class_="bikeDescWrapper")
for i in link:
    j=i.a.text
    link_1.append(j)
print(link_1)
```

```python
pd.DataFrame([link_1]).transpose()
```

```python
link_2=link_1[7]
```

```python
#converting into the csv
with open("web.csv","w") as csv_file:
    write=csv.writer(csv_file)
    write.writerow(image)
    for i in image:
          j=i.img["src"]
          img1.append(j)
    write.writerow(img1)
        
 
    
    
    
```

```python
#opening the file
images=pd.read_csv("web.csv")
```

```python
#getting the price
price=[]
link=soup.findAll("div",class_="text-bold")
for i in link:
    j=i.text[:7]
    price.append(j)
print(price[:7])
```

```python
#updated price list
price_list=price[:7]
```

```python
len(price_list)
```

```python
#Now converting the Bike model and the updated price into the dataFrame
royalEnField=pd.DataFrame({"Royal_model": link_2,
                          "Price":price_list,
                          "Bike_image":royal_image})
```

```python
#viewing the Dataframe that we are created
royalEnField
```

```python
#Now converting what we have extracted from the website into the csv format
royalEnField.to_csv("bikeModel.csv")
```
